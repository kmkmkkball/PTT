```python
import re
import pymongo
from pymongo import MongoClient
client = MongoClient('10.120.27.11', 27017)
db=client["text_mining"]
```

```python
collection = db["ptt_food_<3"]
collection2 = db["ptt_food_<3_ok"]
```

```python
all_list = collection.find()
```

```python
add_sample = ["京都市", "宇治"]
add = []
for all_lists in all_list:
    jin = re.findall("地址：.*(京都市).*",all_lists["內文"].replace(" ",""))
    yu = re.findall("地址：.*(宇治).*",all_lists["內文"].replace(" ",""))
    if len(jin) > 0:
        if jin[0] in add_sample:
            collection2.insert(all_lists)
    if len(yu) > 0:
        if yu[0] in add_sample:
            try:
                collection2.insert(all_lists)
            except Exception as e:
                print(e)
                pass
```
